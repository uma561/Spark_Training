```
# a)Create HR_Employee Pysaprk dataFrame. 
HR=spark.read.csv("dbfs:/FileStore/tables/HR_Employee.csv",header= True ,inferSchema=True)
HR.show()
```

```
#b)	Show printSchema & show 10 rows of Dataset
HR.printSchema()
HR.show(10)
```

```
HR.createOrReplaceTempView("employee")
```

```
#c)	Most employee is working in which department
spark.sql("select Department ,count(*) as most_employee from employee group by Department order by most_employee desc ").show()
```

```
# d)Highest number of job roles
spark.sql("select Max(JobRole) as Job_role,count(*) as most_employee from employee ").show()
```

```
# g)    Mostly hired employee have qualification
spark.sql("select Education, count(*) as Num_Employee from employee  group by Education order by Num_Employee desc").show()
```

```
# e)    Which gender have higher strength as workforce?
spark.sql("select Max(Gender) as Gender,count(*) as Strength from employee").show()
```

```
# h)    Find the count of employee from which education fields
spark.sql("select EducationField, count(*) as Employee from employee group by EducationField order by Employee desc").show()
```

```
 i) What is the job satisfaction level of employee?
spark.sql("select    JobSatisfaction , count(*) as Num_Employee from employee  group by JobSatisfaction order by Num_Employee desc").show()
```

```
# j)    Does most of employee do overtime: Yes or No?
spark.sql("select Overtime, count(*) as Num_Employee from employee where OverTime ='Yes' or Overtime ='No' group by Overtime ").show()
```

```
# k)    Find Min & Max Salaried employees.
spark.sql("select Max(Income) as Max_salary,Min(Income) as Min_salary  from employee ").show()
```

```
# l) Does most of the employee do business travel? Find of the employees counts for each category   BusinessTravel   
spark.sql("select JobRole , count(*) as Num_Employee from employee where BusinessTravel ='Travel_Frequently' group by JobRole ").show()
```

```
# m)    Find the AVG salary of graduate employee.
spark.sql("select mean(Income) as Avg_salary  from employee where Education = 'Bachelor' ").show()
```

```
#n)	Find the employee qualification receiving salary lower than equal to avg. salary of all employee. 
spark.sql("select Income, Education from employee where Income <= (select avg(Income) from employee)").show()
```

```
#o)	When does the employee have highest chance of promotion in terms of working year
spark.sql("select Performance_Rating, count(*) as Promotion from employee group by Performance_Rating order by Promotion desc ").show()
```

```
#p)	Highest attrition is in which department?
spark.sql("select Department, count(*) as totalcount from employee  group by Department order by totalcount desc ").show()
```
